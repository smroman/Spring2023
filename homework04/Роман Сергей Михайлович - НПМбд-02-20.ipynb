6.  Создайте функцию, которая создает $n×n$ матрицу с $(i,j)$-элементом, равным $i+j$.

```python
def ij_sum_matrix(n):
    arr1 = np.zeros((n, n))
    arr2 = np.zeros((n, n))
    for i in range(n):
        for j in range(n):
            arr1[j, i], arr2[i, j] = [i, i]
    return arr1 + arr2

ij_sum_matrix(10)
```
